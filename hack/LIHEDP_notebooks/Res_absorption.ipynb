## Resonant Absorption Ratio

```python
tau = np.linspace(0,2.5,1000)
```

```python
def phi(x):
    return 2.3*x*np.exp(-(2/3)*tau**3)
def A(x):
    return (1/2)*phi(x)**2
```

```python
plt.plot(tau, phi(tau))
plt.xlabel(r'$(k_0 L)^{1/3} \sin \theta$', fontsize=18)
plt.ylabel(r'$\Phi$', fontsize=18)
plt.show()
```

```python
plt.plot(tau, A(tau))
plt.xlabel(r'$(k_0 L)^{1/3} \sin \theta$', fontsize=18)
plt.ylabel(r'$A$', fontsize=18)
plt.show()
```
