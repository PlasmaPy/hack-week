```python
import numpy as np
from sympy.solvers import solve
from sympy import Symbol
from scipy import constants as sc
from matplotlib import pyplot as plt
from decimal import *
```

### Fermi Energy

$$
\boxed{
E_F = \left (\frac{3}{8 \pi} \right )^{2/3} \frac{h^2}{2m_0} n_e^{2/3} 
}
$$ 

```python
(3/(8*sc.pi))**(2/3) * sc.h**2 / (2*sc.m_e) / (sc.eV) * (1e23 * 1e6)**(2/3)
```

$$
\boxed{
E_F \text{ [eV]} = 7.9 \left ( \frac{n_e \text{ [cm$^{-3}$]}}{10^{23}} \right )^{2/3}
}
$$

```python
ns = np.logspace(27,34.7,100)
E_Fs = (3/(8*sc.pi))**(2/3)*sc.h**2/(2*sc.m_e)*ns**(2/3)
Ts = (1/sc.k)*E_Fs
```

```python
plt.plot(np.log10(ns*sc.m_p*1e-3), np.log10(Ts), color='red', linewidth=3)
plt.xlim([-12, 11])
plt.ylim([1.8, 11.2])
plt.show()
```

### Fermi Pressure

$$
\boxed{
P_F = (2/5) n_e E_F = \frac{2}{5} \left (\frac{3}{8 \pi} \right )^{2/3} \frac{h^2}{2m_0} n_e^{5/3}
}
$$

```python
a = (2/5)*(3/(8*sc.pi))**(2/3)*sc.h**2/(2*sc.m_e) / (1e11) * (1e23 * 1e6)**(5/3) 
print(a)
```

$$
\boxed{
P_F \text{ [Mbar]} = 0.5 \left ( \frac{n_e \text{ [cm$^{-3}$]}}{10^{23}} \right )^{5/3} 
}
$$

```python
ne = (1/a)**(3/5) * 1e23 * 1e6 #Corresponds to P_F = 1 Mbar
plt.vlines(x=np.log10(ne*sc.m_p*1e-3), ymin=0, ymax=3.3, color='red', linewidth=3)
plt.xlim([-12, 11])
plt.ylim([1.8, 11.2])
plt.show()
```

### Pressure ionisation

```python
a0 = sc.physical_constants['Bohr radius'][0]
n = 3/(4*sc.pi) * (a0)**(-3)
plt.vlines(x=np.log10(n*sc.m_p*1e-3), ymin=2.5, ymax=4.3, color='red', linewidth=3)
plt.xlim([-12, 11])
plt.ylim([1.8, 11.2])
plt.show()
```
